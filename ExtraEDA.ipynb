```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv")
df.head(5)
```

```python
labels = df['LaunchSite'].unique()
total_launchCount=60
total_launchSuccess=df['LaunchSite'].where(df['Class']==1).value_counts()
sizes = total_launchSuccess/total_launchCount
fig1, ax1 = plt.subplots()
ax1.pie(sizes,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 
plt.legend()
plt.show()
```

```python
labels = ['Success','Fail']
sizes = df['Class'].where(df['LaunchSite']=='CCAFS SLC 40').value_counts()/55
fig1, ax1 = plt.subplots()
ax1.pie(sizes,  labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal') 
plt.legend()
plt.show()
```

```python
sns.scatterplot(y='Class', x="PayloadMass", hue="LaunchSite", data=df.where(df['PayloadMass']<4000))
plt.xlabel("PayloadMass",fontsize=20)
plt.ylabel("Class",fontsize=20)
plt.show()
```
